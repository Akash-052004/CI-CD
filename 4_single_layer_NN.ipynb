```python
#simple feed forward NN

import numpy as np

X = np.array([[1,4], [2,5], [3,6], [4,6]]) # input
y = np.array([[0], [0], [1], [1]]) #op

np.random.seed(42)
weight = np.random.randn(2,1) #(weight, shape)
bias = np.random.randn(1)

def sigmoid(z):
  return 1/ (1+np.exp(-z))

learning_rate = 0.1
epochs = 10

for epoch in range(epochs):
  z = np.dot(X,weight) + bias
  y_pred = sigmoid(z)

  error = y_pred - y #error

  dw = (1 / len(X)) * np.dot(X.T, error) #gradient calculation
  db = (1 / len(X)) * np.sum(error)

  #gradient descenf
  weight = weight - learning_rate * dw
  bias = bias - learning_rate * db

test = np.array([[5,4]])

z = np.dot(test, weight) + bias
y_pred = sigmoid(z)
print(y_pred)

```

```python
if y_pred > 0.5:
  print("1")
else:
  print("0")
```
